## 数据分析
* Feature:属性名
* Unique_values:该属性有多少类属性值
* Percentage of missing values：缺失值占比
* Percentage of values in the biggest category：该类属性中众数占比
* type:数据类型

<font color="blue">value_counts(dropna=False),#计数时将把Null作为单独一种值进行统计</font>

```python
data=[]#分析结果
for col in df_train.columns:
    data.append([col,df_train[col].nunique(),df_train[col].isnull().sum()*100/df_train.shape[0],df_train[col].value_counts(normalize=True,dropna=False).values[0]*100,df_train[col].dtype])
stats_df = pd.DataFrame(data, columns=['Feature', 'Unique_values', 'Percentage of missing values', 'Percentage of values in the biggest category', 'type'])
stats_df.sort_values('Percentage of missing values', ascending=False)[:10]
```

## 特征工程

### 删除只有唯一特性的类别
#### 其中A1、A3、A4暂时不删，后面有需要
删除[A2,A13,A18,A23,B2,B3,B13]

```python
drop_property = list(stats_df[stats_df["Percentage of values in the biggest category"]>90]["Feature"].values)#一类别占比超过百分之90的属性需删掉
drop_property
```

```python
#删除众数占比超过90%的列
df_train=df_train.drop(['A2', 'A13', 'A18', 'A23', 'B2', 'B3', 'B13'],axis=1)
df_test=df_test.drop(['A2', 'A13', 'A18', 'A23', 'B2', 'B3', 'B13'],axis=1)
df_train.shape
df_train.head()
```

### 删除异常值

```python
df_train = df_train[df_train['收率']>0.87]
```

### 区分label,合并test、train

```python
df_train_y = df_train['收率']
df_train_x = df_train.drop(['收率'],axis=1)
df_train_test_x = pd.concat([df_train_x,df_test],axis=0)
df_train_test_x = df_train_test_x.fillna(-1)
```
